Programa 1: Verificar si el numero es par o impar

Programa 3: Pago de un empleado una semana (5 dias), un mes (20 dias)

```
h=int(input("Introduce el numero de horas de trabajo "))
p=150
r1=h*p
r2=h*p*5
r3=h*p*20
print("El salario en un dia es:", r1)
print("El salario en una semana es:", r2)
print("El salario en un mes es:", r3)
```
